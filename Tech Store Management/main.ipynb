```python
import json
def view_products():
    with open("data/products.txt","a+") as pd:
        with open("data/productsindex.txt","r") as id:
            pd.seek(0)
            index = int(id.read())
            if index == 0:
                print("Store is Empty")
            else:
                products = json.load(pd)
                for key,value in products.items():
                    print("\n\nSerial Number: ",key)
                    print("Product Name: ",value[0])
                    print("Description: ",value[1])
                    print("Price: ",value[2])
                    print("Quantity: ",value[3])
def add_products():
    with open("data/products.txt","a+") as pd:
        with open("data/productsindex.txt","a+") as i:
            i.seek(0)
            pd.seek(0)
            index = int(i.read())
            if(index != 0):
                datadict = json.load(pd)
            else:
                datadict = dict()
            listdata = list()
            product_name = input("Enter Product name: ")
            product_description = input("Enter Product Discription: ")
            product_price = input("Enter Product Price: ")
            product_Quantity = input("Enter Quantity: ")
            listdata.append(product_name)
            listdata.append(product_description)
            listdata.append(product_price)
            listdata.append(product_Quantity)
            index += 1
            datadict[index] = listdata
            pd.seek(0)
            pd.truncate()
            json.dump(datadict,pd)
            i.seek(0)
            i.truncate()
            i.write(str(index))
def delete_products():
    with open("data/products.txt","a+") as pd:
        with open("data/productsindex.txt","a+") as i:
            pd.seek(0)
            i.seek(0)
            index = int(i.read())
            if index==0:
                print("Store is Empty")
                return
            while(True):
                view_products()
                num = int(input("Enter Serial Number to delete Product: "))
                if(num>0 and num<=index):
                    datadict = json.load(pd)
                    del datadict[str(num)]
                    index -= 1
                    i.seek(0)
                    i.truncate()
                    i.write(str(index))
                    tempdict = dict()
                    tempindex = 1
                    for key in datadict.keys():
                        templist = datadict[key]
                        tempdict[tempindex] = templist
                        tempindex += 1
                    pd.seek(0)
                    pd.truncate()
                    json.dump(tempdict,pd)
                    print("Successfully deleted Product")
                    break
                else:
                    print("\nInvaid Serial Number")
            
def modify_products():
    with open("data/products.txt","a+") as pd:
        with open("data/productsindex.txt","a+") as i:
            pd.seek(0)
            datadict = json.load(pd)
            view_products()
            i.seek(0)
            index = int(i.read())
            if index == 0:
                print("\nStore is Empty")
            else:
                while True:
                    num = int(input("Enter Serial Number to modify details: "))
                    if(num>0 and num<=index):
                        print("Previous Details: ")
                        print(datadict[str(index)])
                        product_name = input("Enter New Name: ")
                        product_description = input("Enter New Description: ")
                        product_price = input("Enter New Price: ")
                        product_Quantity = input("Enter New Quantity: ")
                        templist = list()
                        templist.append(product_name)
                        templist.append(product_description)
                        templist.append(product_price)
                        templist.append(product_Quantity)
                        datadict[str(index)] = templist
                        pd.seek(0)
                        pd.truncate()
                        json.dump(datadict,pd)
                        print("\nChanges Successfully Done")
                        break
                    else:
                        print("Invalid Syntax")
def manage_product():
    while(True):
        print("1. View Products")
        print("2. Add Products")
        print("3. Delete Products")
        print("4. Modify Products")
        print("5. Exit")
        choice = int(input("Enter choice: "))
        if(choice==1):
            view_products()
        elif(choice==2):
            add_products()
        elif(choice==3):
            delete_products()
        elif(choice==4):
            modify_products()
        else:
            break
def view_sales():
    with open("data/sales.txt","r") as fd:
        with open("data/salesindex.txt","r") as id:
            id.seek(0)
            index = int(id.read())
            if index==0:
                print("No Products Sold")
            else:
                datadict = json.load(fd)
                totalearnings = 0
                print("\n\nProducts Sold: ")
                for key,value in datadict.items():
                    print("\nSale Number: ",key)
                    print("Product Name: ",value[0])
                    print("Quantity: ",value[1])
                    print("Price: ",value[2])
                    totalearnings += int(value[1])*int(value[2])
                print("\nTotal Earnings: ",totalearnings)
def view_employees():
    with open("data/employees.txt","r") as fd:
        with open("data/employeesindex.txt","r") as ed:
            ed.seek(0)
            index = int(ed.read())
            if index == 0:
                print("Currently No Employee")
            else:
                datadict = json.load(fd)
                for key,value in datadict.items():
                    print("Employee Name: ", key)
                    print("Qualification: ",value[0])
                    print("Skills: ",value[1])
def hire_employees():
    with open("data/employees_application.txt","a+") as fd:
        with open("data/employees.txt","a+") as ed:
            with open("data/employees_applicationindex.txt","a+") as id:
                with open("data/employeesindex.txt","a+") as ei:
                    id.seek(0)
                    index = int(id.read())
                    if index == 0:
                        print("No Applications")
                    else:
                        fd.seek(0)
                        application = json.load(fd)
                        while True:
                            print(application)
                            name = input("Enter Employee Name to Accept or Reject Applicaton: ")
                            if name in application.keys():
                                while True:
                                    choice = int(input("Enter 1 to Accept and -1 to Reject: "))
                                    if choice==1:
                                        ei.seek(0)
                                        index = int(ei.read())
                                        if index == 0:
                                            employees = dict()
                                        else:
                                            employees = json.load(ed)
                                        employees[name] = application[name]
                                        ed.seek(0)
                                        ed.truncate()
                                        json.dump(employees,ed)
                                        print("Employee Successfully Accepted")
                                        ei.seek(0)
                                        index = int(ei.read())
                                        index += 1
                                        ei.seek(0)
                                        ei.truncate()
                                        ei.write(str(index))
                                        id.seek(0)
                                        index = int(id.read())
                                        id.seek(0)
                                        id.truncate()
                                        index -= 1
                                        id.write(str(index))
                                        fd.seek(0)
                                        applications = json.load(fd)
                                        del applications[name]
                                        fd.seek(0)
                                        fd.truncate()
                                        json.dump(applications,fd)
                                        break
                                    elif choice==-1:
                                        del appliaction[name]
                                        fd.seek(0)
                                        fd.truncate()
                                        json.dump(application,fd)
                                        print("Employee Successfully Rejecteed")
                                        id.seek(0)
                                        index = int(id.read())
                                        id.seek(0)
                                        id.truncate()
                                        index -= 1
                                        id.write(str(index))
                                        break
                                    else:
                                        print("Invalid Choice, Try Again")
                                break
                            else:
                                print("Invalid Name, Try Again")      
def fire_employees():
    with open("data/employees.txt","a+") as fd:
        with open("data/employeesindex.txt","a+") as ed:
            ed.seek(0)
            index = int(ed.read())
            if index == 0:
                print("Currently No Employee")
            else:
                fd.seek(0)
                employees = json.load(fd)
                while True:
                    name = input("Enter Employee Name to Fire: ")
                    if name in employees.keys():
                        del employees[name]
                        fd.seek(0)
                        fd.truncate()
                        json.dump(employees,fd)
                        ed.seek(0)
                        index -= 1
                        ed.truncate()
                        ed.write(str(index))
                        print("\nEmployee Fired Successfully")
                        break
                    else:
                        print("Invalid Name, Try Again")
def view_applications():
    with open("data/employees_application.txt","a+") as fd:
        with open("data/employees_applicationindex.txt","a+") as id:
            id.seek(0)
            index = int(id.read())
            if index == 0:
                print("No Applications")
            else:
                fd.seek(0)
                data = json.load(fd)
                for key,value in data.items():
                    print("\nEmployee Name: ",key)
                    print("Qualification: ",value[0])
                    print("Skills: ",value[1])
def manage_employees():
    while(True):
        print("1. View Employees")
        print("2. Hire Employees")
        print("3. Fire Employees")
        print("4. View Applications")
        print("5. Exit")
        choice = int(input("Enter choice: "))
        if(choice==1):
            view_employees()
        elif(choice==2):
            hire_employees()
        elif(choice==3):
            fire_employees()
        elif(choice==4):
            view_applications()
        elif(choice==5):
            break
        else:
            print("Invalid choice. TRY AGAIN")
def admin():
    with open("data/password.txt","r+") as fd:
        original_email = fd.readline()
        original_password = fd.readline()
        while(True):
            input_email = input("Enter Email: ")+"\n"
            if(input_email==original_email):
                input_password = input("Enter Password: ")+"\n"
                if(input_password==original_password):
                    while(True):
                        print("1. Manage Products")
                        print("2. View Sales")
                        print("3. Manage Employees")
                        print("4. Exit")
                        choice = int(input("Enter choice: "))
                        if(choice==1):
                            manage_product()
                        elif(choice==2):
                            view_sales()
                        elif(choice==3):
                            manage_employees()
                        elif(choice==4):
                            break
                        else:
                            print("Invalid choice. TRY AGAIN")
                    break
                else:
                    print("Wrong Password")
            else:
                print("Wrong Email")
def buy_product():
    with open("data/products.txt","a+") as fd:
        with open("data/sales.txt","a+") as sd:
            with open("data/productsindex.txt","a+") as i:
                with open("data/salesindex.txt","a+") as si:
                    i.seek(0)
                    index = int(i.read())
                    if index==0:
                        return
                    fd.seek(0)
                    products = json.load(fd)
                    while True:
                        serial = input("Enter Product Number: ")
                        if serial in products.keys():
                            while True:
                                quantity = int(input("Enter Quantity: "))
                                listdata = products[serial]
                                original_quantity = int(listdata[3])
                                if original_quantity==0:
                                    print("Out of Stock")
                                    return
                                if quantity<=(original_quantity):
                                    order_price = (listdata[2])*quantity
                                    si.seek(0)
                                    index = int(si.read())
                                    if index==0:
                                        salesdict = dict()
                                    else:
                                        sd.seek(0)
                                        salesdict = json.load(sd)
                                    saleslist = list()
                                    saleslist.append(listdata[0])
                                    saleslist.append(quantity)
                                    saleslist.append(listdata[2])
                                    listdata[3] = int(listdata[3])-quantity
                                    si.seek(0)
                                    index = int(si.read())
                                    index += 1
                                    si.seek(0)
                                    si.truncate()
                                    si.write(str(index))
                                    sd.seek(0)
                                    sd.truncate()
                                    salesdict[index] = saleslist
                                    json.dump(salesdict,sd)
                                    products[serial] = listdata
                                    fd.seek(0)
                                    fd.truncate()
                                    json.dump(products,fd)
                                    print("\nThanks for Shopping")
                                    break
                                else:
                                    print("Not enough Stock, Try Again")
                            break
                        else:
                            print("Invalid Serial Number")
                    
def customer():
    while(True):
        view_products()
        buy_product()
        breaker = input("Contniue Shopping y/n :")
        if(breaker == "n"):
            break
def employees():
    with open("data/employees_application.txt","a+") as fd:
        with open("data/employees_applicationindex.txt","a+") as ed:
            fd.seek(0)
            ed.seek(0)
            index = int(ed.read())
            if index==0:
                applications = dict()
            else:
                applications = json.load(fd)
            while True:
                print("1. Apply for Job")
                print("2. Withdraw Proposal")
                print("3. Exit")
                choice = int(input("Enter choice: "))
                if choice == 1:
                    name = input("Enter Name: ")
                    qualification = input("Enter Qualifications: ")
                    skills = input("Enter Skills: ")
                    listdata = list()
                    listdata.append(qualification)
                    listdata.append(skills)
                    applications[name] = listdata
                    index += 1
                    ed.seek(0)
                    ed.truncate()
                    ed.write(str(index))
                    fd.seek(0)
                    fd.truncate()
                    json.dump(applications,fd)
                elif choice == 2:
                        name = input("Enter name to withdraw: ")
                        if name in applications.keys():
                            del applicatons[name]
                            index -= 1
                            ed.seek(0)
                            ed.truncate()
                            ed.write(str(index))
                            fd.seek(0)
                            fd.truncate()
                            json.dump(applications,fd)
                            break
                        else:
                            print("No Record Found")
                else:
                    break
def main():
    print("\n\nHaider Tech Store")
    while(True):
        print("1. Admin")
        print("2. Customer")
        print("3. Employees")
        print("4. Exit")
        choice = int(input("Enter choice: "))
        if(choice==1):
            admin()
        elif(choice==2):
            customer()
        elif(choice==3):
            employees()
        elif(choice==4):
            break
        else:
            print("Invalid choice. TRY AGAIN")
if __name__ == "__main__":
    main()
```
